```python
%load_ext autoreload
%autoreload
from lib.transformer import Transformer
from lib.model import Model
from lib.loader import Loader
from lib.predictor import Predictor
from lib.evaluator import Evaluator
```

```python
X, y = Loader().load_data()
```

```python
### hog transformation
X = Transformer(X).hog()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.10, random_state=42)
```

```python
svm_fit = Model(X_train, y_train).svm_classification(kernel='rbf', probability=True,C=1, gamma='scale')
```

```python
pred = Predictor(svm_fit, X_test, True)
y_hat_proba = pred.predict_proba()
y_hat = pred.predict()
```

```python
Evaluator(y_test, y_hat, y_hat_proba).measures()
```

```python
gb_fit = Model(X_train, y_train).gradient_boosting()
pred = Predictor(gb_fit, X_test, True)
y_hat_proba = pred.predict_proba()
y_hat = pred.predict()
```

```python
Evaluator(y_test, y_hat, y_hat_proba).measures()
```

```python
gb_fit = Model(X_train, y_train).gradient_boosting(n_estimators = 500)
pred = Predictor(gb_fit, X_test, True)
y_hat_proba = pred.predict_proba()
y_hat = pred.predict()
```

```python
Evaluator(y_test, y_hat, y_hat_proba).measures()
```
